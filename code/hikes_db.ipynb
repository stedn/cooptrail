```python
from geopy.geocoders import GoogleV3
with open("maps.key", 'r') as f:
    maps_key = f.read()
geolocator = GoogleV3(api_key=maps_key)
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
to_store = []
for item in soup.div.find_all('div',recursive=False):
    name = item.a.text.strip()
    url = item.a['href']
    location = item.find_all('a',class_='styles-module__location___hAqkh')[0].text.strip()

    to_store.append({'name': name,
                     'city': '',
                     'location': location,
                     'search_string': f'{name}, {location}',
                     'url':url})
                            
```

```python
for i in range(len(to_store)):
    to_store[i]['url']='https://www.alltrails.com'+to_store[i]['url']
```

```python
for w in to_store:
    location = geolocator.geocode(w['search_string'])
    if location:
        w['geocode']=location[1]
        w['lat']=location[1][0]
        w['lon']=location[1][1]
```
