```python
from gpx_converter import Converter
import time
import random
```

```python
# trail_list = []
# url_list = []

for ind in range(75,len(amb_list)):
    amb = amb_list[ind]
    website = requests.get(amb)
    soup = BeautifulSoup(website.content, 'html.parser')
    for item in soup.find(class_='ambassador-routes').find_all('div',recursive=False):
        c_split = item.find_all(class_='route-location')[0].text.strip().split(',')
        if (len(c_split)<3):
            continue
        state = c_split[-2]
        city = ' '.join(c_split[:-3])
        url = 'https://ridewithgps.com'+item.a['href']
        
        if url in url_list:
            continue
        url_list.append(url)
        new_website = requests.get(url)
        new_soup = BeautifulSoup(new_website.content, 'html.parser')
        gpx_page = requests.get('https://ridewithgps.com/'+'/'.join(new_soup.find_all('a',text="Send to Device")[0]['href'][1:].split('/')[:2])+'.gpx?sub_format=track')
        if gpx_page.status_code==403:
            print('ruh-roh')
            time.sleep(10)
            continue
        with open(url.split('/')[-1]+".gpx", "wb") as outfile:
            outfile.write(gpx_page.content)
        dic = Converter(input_file=url.split('/')[-1]+".gpx").gpx_to_dictionary()

        trail_list.append({'url':url,
                           'name': item.a.text,
                           'city': city.strip(),
                           'state': state.strip(),
                           'coord': [[dic['longitude'][i],dic['latitude'][i]]for i in range(len(dic['latitude']))]})
        time.sleep(5+random.random()*30)
    time.sleep(5+random.random()*10)
```
