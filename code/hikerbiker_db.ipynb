```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
tst = soup.li
ind = 0
state = 'California'
to_store = []
while tst:
    ind += 1
    url = 'https://www.parks.ca.gov'+tst.a['href']
    name = re.sub(r"[^a-zA-Z0-9 ]", " ", tst.text).strip()
    to_store.append({'name': name,
                     'city': '',
                     'state': state,
                     'search_string': f'{name}, {state}',
                     'url':url})
    tst = tst.next_sibling
                           
```

```python
to_store
```

```python
for w in to_store:
    location = geolocator.geocode(w['search_string'])
    if location:
        w['geocode']=location[1]
        w['lat']=location[1][0]
        w['lon']=location[1][1]
```

```python
import pandas as pd
```

```python
pd.DataFrame(to_store).to_csv('hiker_biker_ca.csv')
```

```python
import json
```

```python
feature_dict = [
    {
      "type": "Feature",
      "properties": {
        "_umap_options": {
          "iconClass": "Drop",
          "iconUrl": "/uploads/pictogram/campsite-24-white.png",
          "color": "DarkGray"
        },
        "description": "",
        "name": f"[[{i['url']}|{i['name']}]]" if 'url' in i else i['name']
      },
      "geometry": {
        "type": "Point",
        "coordinates": [
          i['lon'],
          i['lat']
        ]
      }
    } for i in to_store if 'lat' in i]

with open("hiker_biker_ca.geojson", "w") as outfile:
    json.dump({"type": "FeatureCollection", "features": feature_dict}, outfile, indent=4)
```

```python
souplist = soup.div.find_all('div',recursive=False)
for tst in souplist:
    try:
        name = tst.a.find_all('div',recursive=False)[1].div.div.text
        name = ' '.join(name.replace("\n"," ").strip().split())
        url = "https://stateparks.oregon.gov" + tst.a['href']
        to_store.append({'name': name,
                         'city': '',
                         'state': state,
                         'search_string': f'{name}, {state}',
                         'url':url})  
    except:
        pass
```

```python
to_store = []
state = 'Washington'
for item in soup.div.find_all('div',recursive=False)[1].find_all('div',recursive=False):
    name = item.a.text.strip() 
    url = "https://parks.state.wa.us" + item.a['href']
    to_store.append({'name': name,
                     'city': '',
                     'state': state,
                     'search_string': f'{name}, {state}',
                     'url':url}) 
```

```python
for w in to_store:
    location = geolocator.geocode(w['search_string'])
    if location:
        w['geocode']=location[1]
        w['lat']=location[1][0]
        w['lon']=location[1][1]
```
